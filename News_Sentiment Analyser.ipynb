# Sentiment Article Analysis

##### STOCK ARTICLE SENTIMENT

```python
APPLE_news = pd.read_json ('/Users/ameerirfan/Desktop/FT-BTC-Activities/Project 2/AAPL.json')

df_APPLE = json_normalize(APPLE_news['articles'])

df_APPLE = df_APPLE.drop(['author', 'title', 'url','urlToImage','publishedAt','content','source.id','source.name'], axis=1)

sentiment_title = df_APPLE.apply(lambda x : analyzer.polarity_scores(x['description']), axis=1)

df_APPLE['Comp'] =sentiment_title.apply(lambda x : x.get('compound') )
df_APPLE['Pos'] =sentiment_title.apply(lambda x : x.get('pos') )
df_APPLE['Neu'] =sentiment_title.apply(lambda x : x.get('neu') )
df_APPLE['Neg'] =sentiment_title.apply(lambda x : x.get('neg') )
df_APPLE
```
